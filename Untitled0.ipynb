```python
import pandas as pd
import numpy as np
from sklearn.svm import SVR
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
# load dogecoin data 
from google.colab import files 
files.upload()
```

```python
# read in the data 
df = pd.read_csv('DOGE-USD.csv')
df = df.fillna(0)
# Set the date as index 
df =df.set_index(pd.DatetimeIndex(df['Date'].values))
df
```

```python
future_days = 5
```

```python
# create a new column 
df[str(future_days)+'Day_Price_Forecast'] = df[['Close']].shift(-future_days)
# show the data 
df[['Close',str(future_days)+'Day_Price_Forecast']]
df= df.fillna(0)
df
```

```python
X = np.array(df[['Close']])
X = X[:df.shape[0]- future_days]
print(X)
```

```python
Y = np.array(df[str(future_days)+'Day_Price_Forecast'])
Y = Y[:-future_days]
print(Y)
```

```python
# split the data 

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3)
```

```python

from sklearn.svm import SVR 
svr_rbf = SVR(kernel ='rbf', C = 1e3, gamma = 0.00001)
svr_rbf.fit(x_train, y_train)
```

```python

svr_rbf_confidence = svr_rbf.score(x_test, y_test)
print('svr_rbf_accuracy:',svr_rbf_confidence)
```

```python
svm_prediction = svr_rbf.predict(x_test)
print(svm_prediction)
```

```python
print(y_test)
```

```python
plt.figure(figsize=(14,7))
plt.plot(svm_prediction, label = 'Prediction',lw =2, alpha = 0.7)
plt.plot(y_test, label = 'Actual',lw =2, alpha = 0.7)
plt.title('Prediction vs Actual')
plt.xlabel('Price in USD')
plt.ylabel('Time')
plt.legend(loc = 'upper left')
plt.xticks(rotation =45)
plt.show()
```
